```python
# iterando sobre os produtos necessários
lista_produtos = []

for nome_produto in nomes_produtos:
    time.sleep(3)

    input_place = driver.find_element(By.CLASS_NAME, 'shopee-searchbar-input__input')
    input_place.send_keys(nome_produto)
    input_place.send_keys(Keys.RETURN)

    time.sleep(3)

    # procurando boxes(60 por página)
    try:
        # Espera explícita para a presença dos elementos "boxes"
        boxes_locator = (By.XPATH, "//li[contains(@class, 'col-xs-2-4') and contains(@class, 'shopee-search-item-result__item')]")
        boxes = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located(boxes_locator))
    except Exception as e:
        print("Ocorreu um erro ao obter os elementos 'boxes':", str(e))
        

    k_prod = 0

    for box in boxes:
        k_prod += 1

        try:
            titulo_locator = (By.XPATH, ".//div[@class='ie3A+n bM+7UW Cve6sh']")
            titulo = WebDriverWait(box, 5).until(EC.presence_of_element_located(titulo_locator))
        except (NoSuchElementException, StaleElementReferenceException):
            print(f"Ocorreu um erro na busca do título, produto{k_prod}, na box")
            

        # Espera explícita para a presença do elemento "preco"
        try:
            preco_locator = (By.CLASS_NAME, "ZEgDH9")
            preco = WebDriverWait(box, 5).until(EC.presence_of_element_located(preco_locator))
        except (NoSuchElementException, StaleElementReferenceException):
            print(f"Ocorreu um erro na busca do preço, produto{k_prod}, na box{k_box}")

        # Espera explícita para a presença do elemento "qtde_vendida"
        try:
            qtde_vendida_locator = (By.XPATH, ".//div[@class='r6HknA uEPGHT']")
            qtde_vendida = WebDriverWait(box, 5).until(EC.presence_of_element_located(qtde_vendida_locator))
        except (NoSuchElementException, StaleElementReferenceException):
            print(f"Ocorreu um erro na busca da qtde vendida, produto{k_prod}, na box{k_box}")

        # Espera explícita para a presença do elemento "anuncio"
        
        anuncio_locator = (By.XPATH, ".//div[@class='Sh+UIZ' and @data-sqe='ad']")
        try:
            anuncio = WebDriverWait(box, 5).until(EC.presence_of_element_located(anuncio_locator)).text
        except (AttributeError, NoSuchElementException, IndexError):
            anuncio = None

        # Espera explícita para a presença do elemento "local_envio"
        try:
            local_envio_locator = (By.XPATH, ".//div[@class='zGGwiV']")
            local_envio = WebDriverWait(box, 5).until(EC.presence_of_element_located(local_envio_locator))
        except (NoSuchElementException, StaleElementReferenceException):
            print(f"Ocorreu um erro na busca do local de envio, produto{k_prod}, na box{k_box}")
        
        #Informaçoes adicionais
        titulo.click()
        time.sleep(4)
        
        #esperas explícita para pegar o nome da loja
        try:
            # Aguarde até que o elemento esteja presente no DOM
            nome_loja = (By.CLASS_NAME, "VlDReK")
            WebDriverWait(driver, 10).until(EC.presence_of_element_located(nome_loja))

            # Aguarde até que o elemento esteja visível na página
            WebDriverWait(driver, 10).until(EC.visibility_of_element_located(nome_loja))

            # Obtenha o texto do elemento
            element = driver.find_element(*nome_loja)
            texto_loja = element.text
        except Exception as e:
            print("Ocorreu um erro:", str(e)) 
                      
        driver.back()
        time.sleep(0.5)

        lista_produtos.append([titulo.text, preco.text, qtde_vendida.text, anuncio, local_envio, texto_loja])
```
