````python
import requests
from bs4 import BeautifulSoup
````

````python
url_padrao = 'https://lista.mercadolivre.com.br/'
````

````python
produto_nome = input('Qual produto você deseja?')
````

````python
print(url_padrao + produto_nome)
````

````python
resposta = requests.get(url_padrao+ produto_nome)#o site vai formatar automaticamente- caso nn vc pode formatar
````

````python
print(resposta.text)
````

````python
site = BeautifulSoup(resposta.text, 'html.parser')
````

````python
print(site.prettify())
````

````python
produto =  site.find('div', attrs={'class':  'ui-search-result__wrapper shops__result-wrapper'})
````

````python
print(produto.prettify())
````

#TESTE PESQUISA  INDIVIDUAL

````python
#Titulo,link, preço e vendedor
#class="ui-search-item__title shops__item-title", class="ui-search-price__second-line shops__price-second-line",

titulo= produto.find('h2', attrs={'class': 'ui-search-item__title shops__item-title'})
print('Título do produto: ',titulo.text)
````

````python
preco_symbol = produto.find('span', attrs={'class':'price-tag-symbol'})
preco_fraction = produto.find('span', attrs={'class':'price-tag-fraction'})
preco_cents = produto.find('span',  attrs={'class': 'price-tag-cents'})
preco_formatado = f"{preco_symbol} {preco_fraction},{preco_cents}"
print(f'O preço do {titulo.text} é: {preco_symbol.text} {preco_fraction.text},{preco_cents.text}')


````

````python
link_prod = produto.find('a', attrs={'class': 'ui-search-link'})
print('Link do produto: ', link_prod['href'])
````

````python
#<span class="ui-pdp-color--BLUE ui-pdp-family--REGULAR">HH-DISTRIBUIDORA</span>
vendedor = link_prod.find('span',attrs={'class':'ui-pdp-color--BLUE ui-pdp-family--REGULAR'} )
print(vendedor)
````

# Para pesquisar todos os produtos:

````python
lista_produtos = []
cont = 0
produtos =  site.findAll('div', attrs={'class':  'ui-search-result__wrapper shops__result-wrapper'})

for produto in produtos:
  #titulo
  titulo= produto.find('h2', attrs={'class': 'ui-search-item__title shops__item-title'})
  print('Título do produto: ',titulo.text)

  #preço
  preco_symbol = produto.find('span', attrs={'class':'price-tag-symbol'})
  preco_fraction = produto.find('span', attrs={'class':'price-tag-fraction'})
  preco_cents = produto.find('span',  attrs={'class': 'price-tag-cents'})
  if preco_cents:
    preco_formatado = f"{preco_symbol.text} {preco_fraction.text},{preco_cents.text}"
    print(f'Valor: {preco_formatado}')

  else:
    preco_formatado = f"{preco_symbol.text} {preco_fraction.text}"
    print(f'Valor: {preco_formatado}')


  #link
  link_prod = produto.find('a', attrs={'class': 'ui-search-link'})
  print('Link do produto: ', link_prod['href'])
  cont += 1
  print('Número do produto: ', cont)
  print()

  #lista
  lista_produtos.append([titulo.text, preco_formatado, link_prod['href']])



````

````python
lista_produtos.append([titulo.text, preco_formatado, link_prod['href']])
````
